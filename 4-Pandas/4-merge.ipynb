### merge

```python
import pandas as pd
```

```python
var1 =pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,13,14,15]})
var2 =pd.DataFrame({"A":[1,2,3,4,5],"B":[11,12,43,64,45]})

pd.merge(var1,var2,how ="outer")
```

```python
## Concat

sr1 =pd.Series([1,2,3,4])
sr2 =pd.Series([11,22,33,45])

pd.concat([sr1,sr2])
```

```python
pd.concat([var1,var2])
```
