```python
# groupby

import pandas as pd
```

```python
var =pd.DataFrame({"name":["a","b","c","a","c","b","c","a","c","b","a",],"S_1":[11,4,12,13,14,1,5,1,96,1,5],"S_2":[1,4,2,3,4,16,55,1,6,16,55]})
```

```python
var
```

```python
var_new =var.groupby("name")

var_new
```

```python
for x,y in var_new:
    print(x)
    print(y)
```
