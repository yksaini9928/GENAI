### Advance operation with bussiness exempale

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure : [restaurent_id,2021,2022,2023,2024]
sales_data =np.array([
    [1,1500,1800,2200,2500],  # birani
    [2,1200,1400,1600,1900],  #bites
    [3,2000,2400,2454,2400],#pizza  hub
    [4,1800,2100,2400,2700], #burger point
    [5,1600,1850,2050,2300],# hai point
])
```

### print("=== zomato sales analysis")

```python
print("\n sales data shape",sales_data.shape)
```

```python
print("\n sample data for 1 st 3 rest",sales_data[0:3])
print("\n sample data for 1 st 3 rest",sales_data[:3])

print("\n sample data for 1 st 3 rest",sales_data[:, 1:]) # all colom 
```

```python
# total sales per year
print(np.sum(sales_data,axis=0))
```

```python
yearly_total=np.sum(sales_data[:,1:],axis=0)
print(yearly_total)
```

```python
# minum sales
min_sales =np.min(sales_data[:,1:],axis=1)
print(min_sales)
```

```python
# maximum sales per year
max_sales =np.max(sales_data[:,1:],axis=0)
print(max_sales)
```

```python
# average sale per resturent
avrg_sales =np.mean(sales_data[:,1:],axis=1)
print(avrg_sales)
```

```python
cumsum =np.cumsum(sales_data[:,1:],axis=1)
print(cumsum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("average cumulative sales accros resturent")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
