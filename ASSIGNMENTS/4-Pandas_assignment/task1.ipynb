```python
#pandas Series basic

# Step 1: Import pandas
import pandas as pd
```

```python
# Step 2: Create a Pandas Series
marks = [78, 85, 90, 66, 72]
series = pd.Series(marks)
```

```python
# Step 3: Print the Series
print("Series:")
print(series)
```

```python
# Step 4: Print Series values
print("\nSeries Values:")
print(series.values)
```

```python
# Step 5: Print Index
print("\nIndex:")
print(series.index)
```

```python
# Step 6: Print Data Type
print("\nData Type:")
print(series.dtype)
```

```python
# Step 7: Access the first element
print("\nFirst Element:")
print(series.iloc[0])
```

```python
# Step 8: Access the last two elements
print("\nLast Two Elements:")
print(series.iloc[-2:])
```
