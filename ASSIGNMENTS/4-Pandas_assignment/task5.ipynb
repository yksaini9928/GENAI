```python
import pandas as pd

students={"name":["Amit","Neha","Rahul","Sneha","Pooja"],"Marks":[78,85,90,66,72],"Subject":["Math","Math","Scince","Scince","Math",]}

# convert into data frame

df= pd.DataFrame(students)

df
```

```python
# Information
print("\nInfo:")
print(df.info())

# Description
print("\nDescribe:")
print(df.describe())

# First 5 rows
print("\nHead:")
print(df.head())

# Last 5 rows
print("\nTail:")
print(df.tail())

# Sort by Marks (Descending)
sorted_df = df.sort_values(by='Marks', ascending=False)

print("\nSorted DataFrame:")
print(sorted_df)

# Reset Index
sorted_df = sorted_df.reset_index(drop=True)

print("\nAfter Reset Index:")
print(sorted_df)
```
