```python
import pandas as pd

# Create a Pandas Series
marks = [78, 85, 90, 66, 72]
series = pd.Series(marks)

print("Original Marks:")
print(series)
```

```python
# Find the minimum value
minimum = series.min()

print("Minimum Marks:", minimum)
```

```python
# Find the maximum value
maximum = series.max()

print("Maximum Marks:", maximum)
```

```python
# Find the sum value
sum = series.sum()

print("sum Marks:", sum)
```

```python
# Find the mean value
mean = series.mean()

print("mean Marks:", mean)
```

```python
marks = pd.Series([78, 85, 90, 66, 72])
# Apply lambda function to check pass/fail
result = marks.apply(lambda x: "Pass" if x >= 70 else "Fail")

print("Pass/Fail Status:")
print(result)

# Count students who passed
passed = (marks >= 70).sum()

print("\nNumber of Students Passed:", passed)
```
