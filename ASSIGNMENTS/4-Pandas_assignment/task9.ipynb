```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
sales = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Revenue': [1200, 1500, 900, 2000, 1800]
}

df = pd.DataFrame(sales)

# 1. Total Revenue
print("Total Revenue:", df['Revenue'].sum())

# 2. Average Daily Revenue
avg = df['Revenue'].mean()
print("Average Revenue:", avg)

# 3. Day with Highest Revenue
highest = df.loc[df['Revenue'].idxmax()]
print("\nDay with Highest Revenue:")
print(highest)

# 4. Days where Revenue > Average
print("\nRevenue Above Average:")
print(df[df['Revenue'] > avg])

# 5. Plot Revenue vs Day
df.plot(x='Day', y='Revenue', kind='line', title='Revenue vs Day')
plt.show()
```
