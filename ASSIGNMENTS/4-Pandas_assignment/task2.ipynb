```python
import pandas as pd

# Create a Pandas Series
marks = [78, 85, 90, 66, 72]
series = pd.Series(marks)

print("Original Marks:")
print(series)

# Add 5 marks to all students
series1 = series + 5
series2 = series - 2
series3 = series * 1.05 
series4 = series / 2



print("\n 5 grace Marks:")
print(series1)

print("\nsubtruct 2 Marks:")
print(series2)

print("\n 1.05 multiply Marks:")
print(series3)

print("\ndivide 2 Marks:")
print(series4)
```
