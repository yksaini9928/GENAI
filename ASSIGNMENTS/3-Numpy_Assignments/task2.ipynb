```python
import numpy as np
```

```python
A= np.array([10,20,30,40])
B = np.array([1,2,3,4])

# Addition
C = A + B
print("Addition:", C)   

#subtraction
D = A - B
print("Subtraction:", D)

#multiplication     
E = A * B
print("Multiplication:", E) 

# Division
F = A / B
print("Division:", F)

#power
G = A ** 2
print("Power:", G)

# use numpy function np.add(),np.subtract(),np.multiply(),np.divide(),np.power()
C = np.add(A, B)
D = np.subtract(A, B)
E = np.multiply(A, B)
F = np.divide(A, B)
G = np.power(A, 2)
print("Addition:", C)
print("Subtraction:", D)
print("Multiplication:", E)
print("Division:", F)
print("Power:", G)
```
