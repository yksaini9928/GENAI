```python
#importent numpy mathematical formulas
import numpy as np
```

```python
values = np.array([2,4,6,8,10])
```

```python
# 1 square root of each element
sqrt_values = np.sqrt(values)
print("Square root of each element:", sqrt_values)

```

```python
#2 Exponential of each element
exp_values = np.exp(values)
print("Exponential of each element:", exp_values)
```

```python
# natural logarithm of each element
log_values = np.log(values)     
print("Natural logarithm of each element:", log_values)
```

```python
#4 sum of all elements
sum_values = np.sum(values)
print("Sum of all elements:", sum_values)
```

```python
#5 cumulative sum of elements
cumsum_values = np.cumsum(values)
print("Cumulative sum of elements:", cumsum_values)
```
