```python
# Statistical Operations on NumPy Arrays
import numpy as np
```

```python
marks =np.array([78,85, 90, 66, 72, 88,95,60])
```

```python
#1 mean
mean_marks = np.mean(marks)
print("Mean of marks:", mean_marks)
```

```python
#2 median
median_marks = np.median(marks)
print("Median of marks:", median_marks)
```

```python
#3 satandred deviation
std_dev_marks = np.std(marks)
print(std_dev_marks)
```

```python
# 4 variance
vars =np.var(marks)
print(vars)
```

```python
#5 minimum and maximum
maxi = np.max(marks)
mini =np.min(marks)

print(maxi)
print(mini)
```

```python
range_marks =np.ptp(marks)

print(range_marks)
```
