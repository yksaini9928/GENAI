```python
import numpy as np

sales = np.array([1200, 1500, 900, 2000, 1800, 1700, 1600])

# 1. Total weekly sales
total_sales = np.sum(sales)

# 2. Average daily sales
average_sales = np.mean(sales)

# 3. Highest and lowest sales
highest = np.max(sales)
lowest = np.min(sales)

# 4. Standard deviation
std_dev = np.std(sales)

# 5. Days with sales above average
above_average = sales[sales > average_sales]

print("Total Weekly Sales:", total_sales)
print("Average Daily Sales:", average_sales)
print("Highest Sales:", highest)
print("Lowest Sales:", lowest)
print("Standard Deviation:", std_dev)
print("Sales Above Average:", above_average)
```
