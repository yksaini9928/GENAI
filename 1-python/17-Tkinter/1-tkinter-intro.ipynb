```python
# basic tikinter GUI example
import tkinter as tk

import tkinter.font as tkfont
#TK

window = tk.Tk()
window.title("My First GUI")
window.minsize(400, 300)

custom_font =tk.font.Font(family="Arial", size=24, weight="bold")

label = tk.Label(window, text="Hello, Tkinter!",font=custom_font)
#label.pack(expand=True)
label.pack(padx=20, pady=20)
label.config(fg="blue", bg="yellow")


counter = 0
def function_button():
    global counter
    counter += 1
    label.config(text=f"Button Clicked {counter} times!")

# button = tk.Button(window, text="Click Me!", command=lambda: label.config(text="Button Clicked!"))
button = tk.Button(window, text="Click Me!", command=function_button)# using a function to update the label text
 
button.pack(pady=10)
window.mainloop()
```
